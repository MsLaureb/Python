## Import Data


1. Use the [read_csv()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html) function to read in data from a comma-separated values (csv) file to a Pandas DataFrame called `pie_df`.

```python
#importing pandas as pd
#Reading the csv and assigning the dataframe to pie_df 
import pandas as pd

pie_df = pd.read_csv("JohnnyPiesData.csv")
pie_df

```

## Prepare Data for Linear Regression

1. Drop the `Example` column from the `pie_df` DataFrame, because it offers no information.

2. Encode all categorical data into numeric data via the "One Hot Encoding" technique provided by the Pandas [get_dummies()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html) function.  Display this DataFrame.

3. Since we are performing ordinary least squares linear regression, we will drop one of the newly created Boolean-valued features (output from the `get_dummies()` function) to prevent introducing unwanted correlation in the data.

4. Store the final features in a DataFrame called `features`, and store the positive class labels in a DataFrame called `response`.  Display both of these DataFrames.

**Note:** Since we are not concerned with generalization error in this assignment, we will not split our data into training and test sets. In 'real-world' projects, you would want to split your data to see how your model performs with data that it has never seen before.

```python
#The Example column is dropped because it offers no information
pie_df.drop(['Example'], axis=1, inplace=True)
#View the dataframe
pie_df
```

```python
#One Hot Encoding all the categorical dataframe by using the get_dummies() function
#Store the final features in a DataFrame called features
features = pd.get_dummies(pie_df,columns =['Crust Shape','Crust Size','Crust Shade','Filling Size', 'Filling Shade', 'Class'],drop_first =True)
features # features
```

```python
#Create the labels in a DataFrame called response
response=features[['Class_pos']]
response
```

```python
features =features.drop(['Class_pos'], axis = 1);
features 
```

## Perfrom Linear Regression Model Fitting

1. Import the [LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) class from the `sklearn.linear_model` library. 

2. Instantiate an object of the `LinearRegression` class called `reg_model`.

3. Train the model by invoking the `fit()` method of the `reg_model` object and passing it `features` and `response`.

```python
import numpy as np
# Import the linearRegression class from the sklearn.linear_model library
from sklearn.linear_model import LinearRegression
reg = LinearRegression() #This is a class
#Fit the regression model
reg.fit(features,response)
reg
```

## Examine Linear Regression Model Parameters

1. View the trained model parameters by using the `coef_` and `intercept_` attributes of the trained model.

```python
reg.coef_ # trained model parameters by using the coef_
```

```python
# trained model parameters by using the intercept
reg.intercept_ #This is the y intercept
```

## Making Predictions Using the Linear Regression Model

1. Evaluate the model's performance on the training data set by invoking the `predict()` method and passing `features` to it. 


```python
#Evaluate the model's performance
#It is evelauted on the training data set
reg.predict(features)
```

Below are the results from the linear regression model:

The column "Class_pos" regards the "positive" or negative classification of the pie.  The column "Regression_Predictions" regards the predictions made by the linear regression model directly.  The column "Predicted_Responses" are the adjusted prdeictions made by the model after employing the cut-off values of 0 being 0 <= x <= 0.5 and 1 being 0.5 < x <= 1.0.

```python
import numpy as np

# resp_comp = Response Comparison
#Copies the response object and it is called resp_comp
resp_comp = response.copy() 
#The intertuples iterates through the rows within the features dataframe and assume the next value in the features dataframe
#Index equals False which is considered  and there's not an index colum on the data
#Make a predication using the trained model that wll pass through the data that is inside the feature matrix
#It has to be reshaped therefore it's in the right format
#It is converted into a float and the response is a float
reg_outputs = [float(reg.predict(np.reshape(row, (1, -1)))) for row in features.itertuples(index=False)]
#The List comprehension turns into an array
#It will consist of list of ones in the regression output when wach value is greater than 5
#Otherwise it's going to be zero
# The reg_output has a for loop which is a loop index variable
#It will loop through all the prediction models
#A list of ones in the regression output which is called predicted_resp
predicted_resp = np.array([1 if reg_output >= 0.5 else 0 for reg_output in reg_outputs])
#The reg_output is assigned to a new column called Regression_Predictions
#The predicted_resp is assigned to a new column called Predicted_Responses
#The name of the column is equaled to the data
resp_comp = resp_comp.assign(Regression_Predictions = reg_outputs)
resp_comp = resp_comp.assign(Predicted_Responses = predicted_resp)
resp_comp #The new dataframe
```

## Calculate Model Accuracy

1. Use the [accuracy_score()](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html) function to calculate the accuracy score of the model.

```python
#Calculate the accuracy score by using the SkLearn metrics that is called acccuracy score
from sklearn.metrics import accuracy_score
#Print the accuracy score
#The "accuracy score" is equal to accuracy_score which is a function
#Pass the true labels/values
#Then the predicted values
#Pass the resp_comp with Class_pos and resp_comp with Predicted_Responses 
print("accuracy score", accuracy_score(resp_comp['Class_pos'],resp_comp['Predicted_Responses']))
```
